```python

# Exercise 1: Prime Number Checker

def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, int(num**0.5) + 1):  # Only check up to square root
        if num % i == 0:
            return False
    return True

# Example usage
number = int(input("Enter a number to check if it's prime: "))
if is_prime(number):
    print(f"{number} is a prime number ✅")
else:
    print(f"{number} is not a prime number ❌")
```

```python
# Exercise 2: Product of Random Numbers

import random

# Generate two random numbers between 1 and 10
num1 = random.randint(1, 10)
num2 = random.randint(1, 10)

print(f"Multiply the numbers: {num1} × {num2}")
user_answer = int(input("Enter your answer: "))

if user_answer == num1 * num2:
    print("✅ Correct! Well done.")
else:
    print(f"❌ Wrong! The correct answer is {num1 * num2}")
```

```python
# Exercise 3: Squares of Even Numbers (Chosen: Even numbers)

print("Squares of even numbers between 100 and 200:")

for num in range(100, 201):  # 100 to 200 inclusive
    if num % 2 == 0:  # Only even numbers
        print(f"{num}² = {num ** 2}")
```

```python
# Exercise 4: Hospital Billing System

# Constants
ROOM_CHARGE_PER_DAY = 2000
DOCTOR_FEE = 1500
LAB_TEST_CHARGE = 300
DISCOUNT_THRESHOLD = 10000
DISCOUNT_RATE = 0.10

# Patient details
days_stayed = 4
lab_tests = 3
medicine_charges = 2400

# Calculations
room_charges = days_stayed * ROOM_CHARGE_PER_DAY
lab_charges = lab_tests * LAB_TEST_CHARGE
subtotal = room_charges + DOCTOR_FEE + lab_charges + medicine_charges

# Apply discount if applicable
if subtotal > DISCOUNT_THRESHOLD:
    discount = subtotal * DISCOUNT_RATE
else:
    discount = 0

final_bill = subtotal - discount

# Bill details
print("------ Hospital Bill ------")
print(f"Room charges ({days_stayed} days): ₹{room_charges}")
print(f"Doctor consultation fee: ₹{DOCTOR_FEE}")
print(f"Lab test charges ({lab_tests} tests): ₹{lab_charges}")
print(f"Medicine charges: ₹{medicine_charges}")
print(f"Subtotal: ₹{subtotal}")
print(f"Discount: -₹{discount}")
print(f"Final Bill: ₹{final_bill}")
```
